```python
import gurobipy as gp
from gurobipy import GRB
from datetime import datetime
total_investment = 200000
bond_data = corp_bond

# Define max investment per rating category
max_investment_per_rating = {
    'AAA': 0.3 * total_investment,
    'AA+': 0.3 * total_investment,
    'AA': 0.3 * total_investment,
    'AA-': 0.3 * total_investment,
    'A+': 0.4 * total_investment,
    'A': 0.4 * total_investment,
    'A-': 0.4 * total_investment,
    'BBB+': 0.3 * total_investment,
    'BBB': 0.3 * total_investment,
    'BBB-': 0.3 * total_investment,
    'BB+': 0.2 * total_investment,
    'BB': 0.2 * total_investment,
    'B': 0.2 * total_investment,
    'CCC': 0.1 * total_investment,
    'CC': 0.1 * total_investment,
    'C': 0.1 * total_investment,
    
}

# Create a new model
model = gp.Model("bond_allocation")

# Add decision variables
bond_investments = model.addVars(bond_data.index, vtype=GRB.CONTINUOUS, name="investments")
# Add binary variables for each bond to indicate significant investment
significant_investment = model.addVars(bond_data.index, vtype=GRB.BINARY, name="significant_investment")

# Adjusting the return for a 1 year holding period
holding_period_years = 1
adjusted_returns = bond_data['票面利率(发行时)'] * holding_period_years
# Set the objective: Maximize total adjusted returns
model.setObjective(
    sum(adjusted_returns[bond] * bond_investments[bond] for bond in bond_data.index),
    GRB.MAXIMIZE
)

# Constraints to limit maximum investment in each bond
max_investment_per_bond = 0.1 * total_investment
for bond in bond_data.index:
    model.addConstr(bond_investments[bond] <= max_investment_per_bond, f"max_invest_per_bond_{bond}")
    
# Budget constraint
model.addConstr(sum(bond_investments[bond] for bond in bond_data.index) <= total_investment, "budget")

# Rating scale mapping
rating_scale = {
    'AAA': 17, 'AA+': 16, 'AA': 15, 'AA-': 14, 'A+': 13, 
    'A': 12, 'A-': 11, 'BBB+': 10, 'BBB': 9, 'BBB-': 8, 
    'BB+': 7, 'BB': 6, 'B': 5, 'CCC': 4, 'CC': 3, 'C': 2
}

# Credit rating constraints
for rating, max_invest in max_investment_per_rating.items():
    bonds_in_rating = bond_data[bond_data['债项评级\n[交易日期]2022-06-30\n[债券评级机构]国内评级机构(除中债资信)'] == rating].index
    model.addConstr(sum(bond_investments[bond] for bond in bonds_in_rating) <= max_invest, f"max_invest_{rating}")
    
# Liquidity ratio constraints for each bond
min_liquidity_ratio = 1.5
max_liquidity_ratio = 3.0
latest_liquidity_ratio_col = '流动比率\n[报告期]2022年二季/中报'

for bond in bond_data.index:
    bond_liquidity_ratio = bond_data.loc[bond, latest_liquidity_ratio_col]
    # If the bond's liquidity ratio is outside the range, set investment to zero
    if bond_liquidity_ratio < min_liquidity_ratio or bond_liquidity_ratio > max_liquidity_ratio:
        model.addConstr(bond_investments[bond] == 0, f"liquidity_constraint_{bond}")

# Date constraint: Only invest in bonds with maturity date after 2024-01-01 and issue date before 2022-01-01
maturity_date_threshold = datetime(2024, 1, 1)
# issue_date_threshold = datetime(2021, 1, 1)

for bond in bond_data.index:
    maturity_date = datetime.strptime(bond_data.loc[bond, '到期日期'], '%Y-%m-%d')
    # issue_date = datetime.strptime(bond_data.loc[bond, '起息日期'], '%Y-%m-%d')
    # if maturity_date <= maturity_date_threshold or issue_date >= issue_date_threshold:
    #     model.addConstr(bond_investments[bond] == 0, f"date_constraint_{bond}")
    if maturity_date <= maturity_date_threshold:
        model.addConstr(bond_investments[bond] == 0, f"date_constraint_{bond}")
 


# Rating decline constraint 
rating_columns = [
    '债项评级\n[交易日期]2021-12-31\n[债券评级机构]国内评级机构(除中债资信)',
    '债项评级\n[交易日期]2022-06-30\n[债券评级机构]国内评级机构(除中债资信)'
]

for bond in bond_data.index:
    rating_previous = rating_scale[bond_data.loc[bond, rating_columns[0]]]
    rating_current = rating_scale[bond_data.loc[bond, rating_columns[1]]]
    if rating_previous > rating_current+3:
        model.addConstr(bond_investments[bond] == 0, f"rating_decline_constraint_{bond}")

# only invest the bond rate greater than BBB+
min_allowed_rating = rating_scale['AA']
latest_rating_column = '债项评级\n[交易日期]2022-06-30\n[债券评级机构]国内评级机构(除中债资信)'  # Replace with your actual column name

decline_threshold = 3  # Set the threshold for rating decline

for bond in bond_data.index:
    rating_previous = rating_scale[bond_data.loc[bond, rating_columns[0]]]
    rating_current = rating_scale[bond_data.loc[bond, rating_columns[1]]]
    if rating_previous - rating_current > decline_threshold:
        model.addConstr(bond_investments[bond] == 0, f"rating_decline_constraint_{bond}")  
        
# Constraint for significant investments
min_bonds_to_invest = 70
significant_investment_threshold = 0.01 * total_investment
for bond in bond_data.index:
    model.addConstr(bond_investments[bond] >= significant_investment_threshold * significant_investment[bond], f"link_sig_inv_{bond}")
model.addConstr(sum(significant_investment[bond] for bond in bond_data.index) >= min_bonds_to_invest, "min_num_bonds")

model.optimize()
```

```python
# Retrieve and print the investment plan
investment_plan = model.getAttr('x', bond_investments)
for bond, investment in investment_plan.items():
    print(f"Bond {bond}: Invested amount = {investment}")

# Sum up the total invested amount
total_invested_amount = sum(investment_plan.values())
print(f"Total Invested Amount: {total_invested_amount}")
```

```python
# Calculate total returns
total_returns = sum(bond_data.loc[bond, '票面利率(发行时)']/100 * investment_plan[bond] for bond in bond_data.index)

# Calculate the total invested amount
total_invested_amount = sum(investment_plan.values())

# Calculate the return rate
if total_invested_amount > 0:
    return_rate = total_returns / total_invested_amount
else:
    return_rate = 0

print(f"Total Returns: {total_returns}")
print(f"Total Invested Amount: {total_invested_amount}")
print(f"Return Rate: {return_rate}")
```
